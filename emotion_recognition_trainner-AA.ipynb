mported the required libraries

```python
import sys, os
import pandas as pannd
import numpy as nummp
```

```python
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D, BatchNormalization,AveragePooling2D
from keras.losses import categorical_crossentropy
from keras.optimizers import Adam
from keras.regularizers import l2
from keras.utils import np_utils
```

ataset is in the csv file

```python
df=pannd.read_csv('fer2013.csv')
```

hese are the array which are classified

```python
X_train,train_y,X_test,test_y=[],[],[],[]
```

```python
for index, ro in df.iterrows():
    val=ro['pixels'].split(" ")
    try:
        if 'Training' in ro['Usage']:
            X_train.append(nummp.array(val,'float32'))
            train_y.append(ro['emotion'])
        elif 'PublicTest' in ro['Usage']:
            X_test.append(nummp.array(val,'float32'))
            test_y.append(ro['emotion'])
    except:
        print(f"error occuring in index :{index} and row:{ro}")
```

```python
num_features = 64
num_labels = 7
batch_size = 64 #how many images to be trained at once
epochs = 1 #for the processing , period
#size of the image
width, height = 48, 48
```

loat acceptiong a parameter that is only one

```python
X_train = nummp.array(X_train,'float32')
train_y = nummp.array(train_y,'float32')
X_test = nummp.array(X_test,'float32')
test_y = nummp.array(test_y,'float32')
```

or the conversion of class vector into a matrices

```python
train_y=np_utils.to_categorical(train_y, num_classes=num_labels)
test_y=np_utils.to_categorical(test_y, num_classes=num_labels)
```

annot produce<br>
or the normalization of data between 0 and 1<br>
ean algorith<br>
tandard deviation algorithm

```python
X_train -= nummp.mean(X_train, axis=0)
X_train /= nummp.std(X_train, axis=0)
```

```python
X_test -= nummp.mean(X_test, axis=0)
X_test /= nummp.std(X_test, axis=0)
```

```python
X_train = X_train.reshape(X_train.shape[0], 48, 48, 1)
```

```python
X_test = X_test.reshape(X_test.shape[0], 48, 48, 1)
```

design of the convolutional neural network<br>
umber 1 convolution layer

```python
modl = Sequential()
```

```python
modl.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(X_train.shape[1:])))
modl.add(Conv2D(64,kernel_size= (3, 3), activation='relu'))
# model.add(BatchNormalization()) # for the normalization
modl.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))
modl.add(Dropout(0.5))
```

umber 2 convolution layer

```python
modl.add(Conv2D(64, (3, 3), activation='relu'))
modl.add(Conv2D(64, (3, 3), activation='relu'))
# model.add(BatchNormalization()) # for the normalization
modl.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))
modl.add(Dropout(0.5))
```

umber 3 convolution layer

```python
modl.add(Conv2D(128, (3, 3), activation='relu'))
modl.add(Conv2D(128, (3, 3), activation='relu'))
# model.add(BatchNormalization()) # for the normalization
modl.add(MaxPooling2D(pool_size=(2,2), strides=(2, 2)))
```

```python
modl.add(Flatten())
```

or the full connection of the neural networks<br>
ectifies linear unit for the activation

```python
modl.add(Dense(1024, activation='relu'))
modl.add(Dropout(0.2))
#relu directly outputs the input 
modl.add(Dense(1024, activation='relu'))
modl.add(Dropout(0.2))
```

oftmax for the prediction of the multinomial distribution

```python
modl.add(Dense(num_labels, activation='softmax'))
```

summary of model<br>
he following function compiles the models

```python
modl.compile(loss=categorical_crossentropy,
              optimizer=Adam(),
              metrics=['accuracy']) #for true results
```

he following trains the models

```python
modl.fit(X_train, train_y,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1, #program prolix
          validation_data=(X_test, test_y),
          shuffle=True)
```

aves the model for later use

```python
fer_json = modl.to_json()
with open("fer.json", "w") as json_file: 
    json_file.write(fer_json)
modl.save_weights("fer.h5")
```
